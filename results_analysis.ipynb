## Richocet Dirty Entities
### Detailed
`TOKEN_BIGRAMS_TF_IDF`

```python
columns = ['elapsed_time', 'window_size', 'filtering_threshold', 'representation_model', 
           'similarity_metric', 'group_linkage_threshold', 'ricochet_threshold', 'precision', 'recall', 'f_score']
```

```python
d1 = pd.read_csv('results/dirty-gs-w2-3-001-015.csv', names=columns)
d2 = pd.read_csv('results/dirty-gs-w5-15-001-015.csv', names=columns)
detailed = d1.append(d2)
```

```python
detailed.head()
```

```python
detailed.describe()
```

```python
detailed[(detailed.window_size == 2) & (detailed.filtering_threshold == 0.1)].pivot(index='group_linkage_threshold', columns='ricochet_threshold', values='f_score')
```

```python
pd.pivot_table(detailed, index='window_size', columns='filtering_threshold', values='elapsed_time')
```

```python
pd.pivot_table(detailed, index='window_size', columns='filtering_threshold',
               values=['precision', 'recall', 'f_score'])
```

```python
pd.pivot_table(detailed[(detailed.window_size == 2) & (detailed.filtering_threshold == 0.1)], 
               index='group_linkage_threshold', columns='ricochet_threshold', 
               values=['precision', 'recall', 'f_score'])
```

```python
tmp = pd.pivot_table(detailed[(detailed.ricochet_threshold <=0.5) & (detailed.group_linkage_threshold <=0.5)], 
                     columns=['ricochet_threshold','group_linkage_threshold'], 
                     index=['window_size','filtering_threshold'], 
                     values=['precision', 'recall', 'f_score'])
tmp.to_latex('tex/dirty-ricochet-detailed-result-all.tex', caption='Ricochcet - dirty all results dtailed', float_format="%.2f")
tmp
```

```python
for label in ['precision', 'recall', 'f_score']:
    t = pd.pivot_table(detailed[(detailed.ricochet_threshold <=0.5) & (detailed.group_linkage_threshold <=0.5)], 
                       columns=['ricochet_threshold','group_linkage_threshold'], 
                       index=['window_size','filtering_threshold'], values=label)
    t.to_latex('tex/dirty-dirty-detailed-'+label+'-all.tex', caption='Ricochet - detailed dirty - '+label, float_format="%.2f")
    print('-----', label, '-----')
    ICD.display(t.style.highlight_max(axis=0).highlight_max(axis=1, color='green'))
```
