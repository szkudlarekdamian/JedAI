# Results analysis

```python
from IPython.core import display as ICD

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Unique Mapping Clustering
`TOKEN_BIGRAMS_TF_IDF`

```python
umc_bi = pd.read_csv('results/grid-search-cc-unique-mapping-clustering-bi.csv')
```

```python
for label in ['precision', 'recall', 'f_score']:
    t = pd.pivot_table(umc_bi, 
                       columns=['clustering_threshold','group_linkage_threshold'], 
                       index=['window_size','filtering_threshold'], values=label)
    t.to_latex('tex/unique-mapping-clustering-'+label+'-all.tex', caption='Unique Mapping Clustering - '+label, float_format="%.2f")
    ICD.display(t.style.highlight_max(axis=0).highlight_max(axis=1, color='green'))
```

```python
tmp = pd.pivot_table(umc_bi, index='window_size', columns='filtering_threshold', values='elapsed_time')
tmp.to_latex('tex/unique-mapping-clustering-time-bigrams.tex', caption='Unique Mapping Clustering - bigrams - execution time [ms]', float_format="%.2f")
tmp
```
